```python
import numpy as np
import scipy
import scipy.misc
import filter as f
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
im = scipy.misc.lena();
im = np.double(im) / 255;
```

```python
k1 = np.array([[1,0,-1],[2,0,-2],[1,0,-1]],dtype=np.double)
k2 = np.array([[1,2,1],[0,0,0],[-1,-2,-1]],dtype=np.double)
```

```python
ed1 = f.Filt(im, k1, 0)
ed2 = f.Filt(im, k2, 0)
ed = (ed1**2 + ed2**2)**0.5
fig = plt.figure()
fig.add_subplot(1,2,1)
plt.imshow(im, cmap='gray')
fig.add_subplot(1,2,2)
plt.imshow(ed[:,:,0],cmap='gray');
plt.show()
```
