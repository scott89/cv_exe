```python
import numpy as np
import scipy
import scipy.misc
import PyCV as f
import matplotlib.pyplot as plt
import scipy.ndimage
%matplotlib inline
```

```python
im = scipy.misc.lena();
im = np.double(im) / 255;
lbp = f.ExtractLBP(im, 0, [16, 2])
```

```python
fig = plt.figure();
fig.set_size_inches([1,1]);
ax1 = plt.axes([0,0,4,4]);
ax1.set_axis_off();
fig.add_axes(ax1);
plt.imshow(im,cmap='gray')
plt.title('Input image')
ax2 = plt.axes([5,0,4,4]);
ax2.set_axis_off();
fig.add_axes(ax2);
plt.imshow(lbp[:,:,0],cmap='gray')
plt.title('LBP_uri(16,2) feature')
plt.show()
```

```python
h=np.histogram(lbp,16);
plt.bar(h[1][0:-1],h[0])
plt.title('Histogram of LBP code');
plt.show()
```
